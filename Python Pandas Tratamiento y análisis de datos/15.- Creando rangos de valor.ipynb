```python
import pandas as pd
```

```python
datos = pd.read_csv('alquiler.csv', sep = ';')
datos.head(10)
```

```python
# 1 y 2
# 3 y 4
# 5 y 6
# 7 o mas cuartos 
```

```python
clases = [0, 2, 4, 6, 100]
```

```python
cuartos = pd.cut(datos.Cuartos, clases)
cuartos
# Recorre cada elemento y le asigna la categoria
```

```python
pd.value_counts(cuartos)
# Muestra las cantidades que hay en cada clasificacion
```

```python
label = ['1 y 2 cuartos', '3 y 4 cuartos', '5 y 6 cuartos', '7 cuartos o mas']
# Agregar labels para cada clasificacion
```

```python
cuartos = pd.cut(datos.Cuartos, clases, labels = label)
cuartos
# Agregar labels a la serie
```

```python
pd.value_counts(cuartos)
```

```python
cuartos = pd.cut(datos.Cuartos, clases, labels = label, include_lowest = True)
# Incluir los que no estan en la llave (con valor 0)
```

```python
pd.value_counts(cuartos)
```
