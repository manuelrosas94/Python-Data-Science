## Inmuebles residenciales

```python
import pandas as pd
```

```python
datos = pd.read_csv('alquiler.csv', sep = ';')
datos.head(10)
```

```python
list(datos['Tipo'].drop_duplicates())
```

```python
residencial = ['Habitación',
 'Casa',
 'Departamento',
 'Casa en condominio',
 'Casa comercial',
 'Casa de villa']
residencial
# Seleccion de inmuebles habitables
```

```python
datos.head(10)
```

```python
seleccion = datos['Tipo'].isin(residencial)
seleccion
# Comparar si nuestra seleccion esta en la tabla original
```

```python
datos_residencial = datos[seleccion]
datos_residencial
# Dataframe con la la nueva variable que queremos visualizar en la tabla
```

```python
list(datos_residencial['Tipo'].drop_duplicates())
```

```python
datos_residencial.shape[0]
```

```python
datos.shape[0]
```

```python
datos_residencial.index = range(datos_residencial.shape[0])
# Convertir una secuencia de filas sin saltar numeros
```
