## Tratamiento de datos faltantes(continuacion)

```python
datos
```

```python
datos[datos['Mantenimiento'].isnull()].shape[0]
```

```python
seleccion = (datos['Mantenimiento'].isnull()) & (datos['Tipo'] == 'Departamento')
# Criterio de seleccion asignado a una variable
```

```python
A = datos.shape[0]
datos = datos[~seleccion]
B = datos.shape[0]
A - B
# Valores que se eliminaron de la seleccion
```

```python
datos[datos['Mantenimiento'].isnull()].shape[0]
# Cantidad de nulos en Mantenimiento
```

```python
datos = datos.fillna({'Mantenimiento':0, 'Impuesto':0})
# Nuevo dataframe rellenando mantenimiento e impuesto
```

```python
datos[datos['Mantenimiento'].isnull()].shape[0]
```

```python
datos[datos['Impuesto'].isnull()].shape[0]
```

```python
datos.info()
```

```python
datos
```

```python
datos.to_csv('alquiler_residencial.csv', sep = ';', index = False)
```

# Ejercicio

```python
inmuebles= pd.DataFrame([['Departamento', None, 970, 68], 
                        ['Departamento', 2000, 878, 112], 
                        ['Casa', 5000, None, 500], 
                        ['Departamento', None, 1010, 170], 
                        ['Departamento', 1500, 850, None], 
                        ['Casa', None, None, None], 
                        ['Departamento', 2000, 878, None], 
                        ['Departamento', 1550, None, 228], 
                        ['Departamento', 2500, 880, 195]], 
                        columns = ['Tipo', 'Valor', 'Mantenimiento', 'Impuesto'])
inmuebles
```

```python
# Elimina los registros que tengan la columna Valor nula
```

```python
inmuebles.dropna(subset = ['Valor'], inplace = True)
inmuebles
```

```python
#  Elimina los inmuebles de tipo Departamento que tengan la columna Mantenimiento nula
```

```python
seleccion = ((inmuebles['Mantenimiento'].isnull()) & inmuebles['Tipo'] == 'Departamento')
inmuebles= inmuebles[~seleccion]
inmuebles
```

```python
# Reemplaza los valores nulos que quedan en las variables Mantenimiento y Impuesto por cero
```

```python
inmuebles = inmuebles.fillna({'Mantenimiento': 0, 'Impuesto': 0})
inmuebles
```

```python
# Reconstruir el indice del DataFrame resultante
```

```python
inmuebles.index = range(inmuebles.shape[0])
inmuebles
```
