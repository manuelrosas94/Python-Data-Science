```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (15, 10))
```

```python
datos = pd.read_csv('alquiler_residencial_sin_outlier.csv', sep = ';')
datos.head(10)
```

```python
area = plt.figure()
```

```python
g1 = area.add_subplot(2,2,1)
g2 = area.add_subplot(2,2,2)
g3 = area.add_subplot(2,2,3)
g4 = area.add_subplot(2,2,4)
# Determina el tamaño de la grafica y la posicion de cada subgrafica
```

```python
# Crear graficos
g1.scatter(datos.Valor, datos.Area)
g1.set_title('Valor x Area')

g2.hist(datos.Valor)
g2.set_title('Histograma')

datos_g3 = datos.Valor.sample(100)
datos_g3.index = range(datos_g3.shape[0])
g3.plot(datos_g3)
g3.set_title('Muestra (Valor)')

grupo = datos.groupby('Tipo')['Valor']
label = grupo.mean().index
valores = grupo.mean().values
g4.bar(label, valores)
g4.set_title('Valor Medio por Tipo')

# Crea un grafico con los valores asignados
```

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (15, 7))

datos_muestra = pd.read_csv('alquiler_muestra.csv', sep = ';')
```

```python
area = plt.figure()
g1 = area.add_subplot(1, 2, 1)
g2 = area.add_subplot(1, 2, 2)
grupo1 = datos_muestra.groupby('Tipo Agrupado')['Valor']
label = grupo1.count().index
valores = grupo1.count().values
g1.pie(valores, labels = label, autopct='%1.1f%%')
g1.set_title('Total de Inmuebles por Tipo Agrupado')
grupo2 = datos_muestra.groupby('Tipo')['Valor']
label = grupo2.count().index
valores = grupo2.count().values
g2.pie(valores, labels = label, autopct='%1.1f%%', explode = (.1, .1, .1, .1, .1))
g2.set_title('Total de Inmuebles por Tipo')
```
