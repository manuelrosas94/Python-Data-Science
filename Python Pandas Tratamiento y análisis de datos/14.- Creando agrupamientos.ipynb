# Ejercicio

```python
import pandas as pd
alumnos = pd.DataFrame({
'Nombre': ['Ary', 'Katia', 'Denis', 'Beto', 'Bruna', 'Dara', 'Carlos', 'Alice'], 
'Sexo': ['M', 'F', 'M', 'M', 'F', 'F', 'M', 'F'], 
'Edad': [15, 27, 56, 32, 42, 21, 19, 35], 
'Notas': [7.5, 2.5, 5.0, 10, 8.2, 7, 6, 5.6], 
'Aprobado': [True, False, False, True, True, True, False, False]}, 
columns = ['Nombre', 'Edad', 'Sexo', 'Notas', 'Aprobado'])
alumnos
```

```python
# ¿Cómo debemos proceder para obtener un DataFrame con las notas promedio de los alumnos, con dos casas decimales, agrupado por su sexo?
```

```python
sexo = alumnos.groupby('Sexo')
sexo = pd.DataFrame(sexo['Notas'].mean().round(2))
sexo.columns = ['Notas Médias']
sexo
```

# Estadisticas descriptivas

```python
grupo_barrio['Valor'].describe().round()
# Muestra estadisticas de la columna seleccionada
```

```python
grupo_barrio['Valor'].aggregate(['min', 'max']).rename(columns = {'min':'Minimo', 'max': 'Maximo'})
# aggregate Extrae estadisticas que queremos visualizar
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (20,10))
```

```python
fig = grupo_barrio['Valor'].mean().plot.bar(color = 'blue')
fig.set_ylabel('Valo del alquiler')
fig.set_title('Valor Medio del Alquiler por Distrito', {'fontsize':22})
```

```python
fig = grupo_barrio['Valor'].max().plot.bar(color = 'blue')
fig.set_ylabel('Valor del alquiler')
fig.set_title('Valor Maximo del Alquiler por Distrito', {'fontsize':22})
```

# Ejercicio

```python
precios = pd.DataFrame([
['Feria', 'Cebolla', 2.5], 
         ['Mercado', 'Cebolla', 1.99], 
         ['Supermercado', 'Cebolla', 1.69], 
         ['Feria', 'Tomate', 4], 
         ['Mercado', 'Tomate', 3.29], 
         ['Supermercado', 'Tomate', 2.99], 
         ['Feria', 'Papa', 4.2], 
         ['Mercado', 'Papa', 3.99], 
         ['Supermercado', 'Papa', 3.69]], 
columns = ['Local', 'Producto', 'Precio'])
precios
```

```python
productos = precios.groupby('Producto')
productos.describe().round(2)
```

```python
estadisticas = ['mean', 'std', 'min', 'max']
nombres = {'mean': 'Promedio', 'std': 'Desviación Estándar', 
    'min': 'Mínimo', 'max': 'Máximo'}
productos['Precio'].aggregate(estadisticas).rename(columns = nombres).round(2)
```
