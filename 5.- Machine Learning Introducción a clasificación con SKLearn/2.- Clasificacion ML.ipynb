```python
import pandas as pd
uri = "https://gist.githubusercontent.com/guilhermesilveira/2d2efa37d66b6c84a722ea627a897ced/raw/10968b997d885cbded1c92938c7a9912ba41c615/tracking.csv"
datos = pd.read_csv(uri)
datos.head()
```

```python
x = datos[[	'home',	'how_it_works',	'contact']]
y = datos['bought']
```

```python
nuevos_nombres = {
    'home': 'Inicio',	
    'how_it_works': 'Como funciona',	
    'contact': 'Contacto', 
    'bought': 'Compra'
}
```

```python
datos = datos.rename(columns = nuevos_nombres)
datos.head()
```

```python
x = datos[['Inicio',	'Como funciona',	'Contacto']]
y = datos['Compra']
datos.shape
```

```python
train_x = x[:75]
train_y = y[:75]
test_x = x[75:]
test_y = y[75:]
train_x.shape
# Delimitar cuantos datos son para entrenar y cuantos son para probar
```

```python
from sklearn.svm import LinearSVC
modelo = LinearSVC()
modelo.fit(train_x, train_y)
```

```python
from sklearn.metrics import accuracy_score
pred = modelo.predict(test_x)
accuracy = accuracy_score(test_y, pred) * 100

print("Porcentaje de acieto %.2f " % (accuracy))
```

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

SEED = 20
train_x, test_x, train_y, test_y= train_test_split(x, y, test_size=0.25, random_state=SEED, stratify=y)
modelo = LinearSVC()
modelo.fit(train_x, train_y)

pred = modelo.predict(test_x)
accuracy = accuracy_score(test_y, pred) * 100

print("Porcentaje de acieto %.2f " % (accuracy))
```

```python
# SEED indica apartir de donde empieza a tomar las muestras
# train_test_split toma de forma aleatoria las muestras para entrenar y para probar
# stratify estratifica el parametro asignado, para mantener una proporcion de 1:2 los train y test
```
