```python
# Features
# Pelo largo ?, pata corta?, ladra?
# 1:Si   0: No

cerdito1 = [0,1,0]
cerdito2 = [1,1,0]
cerdito3 = [0,1,1]

perro1 = [0,1,1]
perro2 = [1,0,1]
perro3 = [1,1,1]

train_x = [cerdito1, cerdito2, cerdito3, perro1, perro2, perro3]
train_y = [0,0,0,1,1,1] # 0:Cerdito, 1:Perro
print(train_x)
```

```python
from sklearn.svm import LinearSVC
modelo = LinearSVC()
modelo.fit(train_x, train_y)
```

```python
misterioso = [1,1,1]
misterioso2 = [1,1,0]
misterioso3 = [0,1,1]

test_x = [misterioso, misterioso2, misterioso3]
test_y = [1,0,1]

pred = modelo.predict(test_x)

correctos = (pred == test_y).sum()
tot = len(test_y)
prop_aciertos = correctos/tot

print("% de acieto", prop_aciertos*100)
```

```python
from sklearn.metrics import accuracy_score
print("% de acierto", accuracy_score(test_y, pred) * 100)
print("Porcentaje de acieto %.2f " % (accuracy_score(test_y, pred) * 100))
# accuracy_score(test que queremos evaluar, prediccion que arrojo el modelo)
```
