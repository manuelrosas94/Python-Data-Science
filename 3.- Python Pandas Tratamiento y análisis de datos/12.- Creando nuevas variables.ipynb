## Creando nuevas variables

```python
import pandas as pd
```

```python
datos = pd.read_csv('alquiler_residencial.csv', sep = ';')
# Usando el csv sobre escrito de la actividad 10
```

```python
datos.head(10)
```

```python
datos['Valor Bruto'] = (datos['Valor']) + (datos['Mantenimiento']) + (datos['Impuesto'])
datos.head(10)
# Crea una nueva columna al dataframe con los valores elegidos
```

```python
datos['Valor m2'] = (datos['Valor']) / (datos['Area'])
datos.head(10)
```

```python
datos['Valor m2'] = datos['Valor m2'].round(2)
datos.head(10)
# Redondear el valor con 2 decimales
```

```python
datos['Valor Bruto m2'] = ((datos['Valor Bruto']) / (datos['Area'])).round(2)
datos.head(10)
```

```python
casa = ['Casa', 'Casa en condominio', 'Casas de villa']
```

```python
datos['Agrupado'] = datos['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Departamento')
datos
# Usando lambda para clasificar en casa o departamento
```
