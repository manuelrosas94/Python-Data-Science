```python
import pandas as pd
```

```python
data = [0.5, None, None, 0.52, 0.54, None, None, 0.59, 0.6, None, 0.7]
s = pd.Series(data)
s
```

```python
s.fillna(0)
```

```python
s
```

```python
s.fillna(method='ffill')
# Rellena el valor nulo con el valor anterior que no es nulo hasta teminar
```

```python
s.fillna(method= 'bfill')
# Empieza desde el ultimo elemento hasta el primero
```

```python
s
```

```python
s.fillna(s.mean())
# Rellena los valores nulos por la media de los no nulos
```

```python
s
```

```python
s.fillna(method= 'ffill', limit = 1)
# El limit solo rellena las veces que se especifican
```

```python
s
```

```python
s1 = s.fillna(method= 'ffill', limit = 1)
s1
# Rellenar la primera parte con ffill
```

```python
s2 = s1.fillna(method= 'bfill', limit = 1)
s2
# Rellena la ultima parte con bfill
```
