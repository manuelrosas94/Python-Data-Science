# Reporte analisis IV

Selecciones y frecuencias

```python
import pandas as pd
```

```python
datos = pd.read_csv('alquiler_residencial.csv', sep = ';')
datos.head(10)
```

```python
#Seleccione solamente los inmuebles clasificados con tipo'Departamento"
```

```python
seleccion = datos['Tipo'] == 'Departamento'
n1 = datos[seleccion].shape[0]
n1
# shape[] solo arroja el numero del dataframe en la posicion seleccionada
```

```python
#Seleccione los inmuebles clasificados con tipos'Casa','Casa en condominio'y'Casa de villa'
```

```python
seleccion = (datos['Tipo'] == 'Casa') | (datos['Tipo'] == 'Casa en condominio') | (datos['Tipo'] == 'Casa de villa')
n2 = datos[seleccion].shape[0]
n2
```

```python
#Seleccione los inmuebles con area entre 60 y 100 metros cuadrados, incluyendo los limites
```

```python
seleccion = (datos['Area'] >= 60) & (datos['Area'] <= 100)
n3 = datos[seleccion].shape[0]
n3
```

```python
# Seleccione los inmuebles que tengan por lo menos 4 cuartos y alquiler menor que $2000.00
```

```python
seleccion = (datos['Cuartos'] >= 4) & (datos['Valor'] < 2000)
n4 = datos[seleccion].shape[0]
n4
```

```python
print("No. de inmuebles clasificados con tipo 'Departamento' -> {}".format(n1))
print("No. de inmuebles clasificados con tipos 'Casa', 'Casa en condominio' y 'Casa en Villa' -> {}".format(n2))
print("No. de inmuebles con area entre 60 y 100 metros cuadrados, incluyendo los limites -> {}".format(n3))
print("No. de inmuebles que tengan por lo menos 4 cuartos y alquiler menor que $2000.00 -> {}".format(n4))


```
