```python
import pandas as pd
```

# Series


```python
data = [1,2,3,4,5]
```

```python
s = pd.Series(data)
s
```

```python
index = ['Linea' + str(i) for i in range(5)]
index
# Nombres de los nuevos indices
```

```python
s = pd.Series(data = data, index = index)
s
# Asigna los nuevos indices a los valores
```

```python
data = {'Linea' + str(i) : i+1 for i in range(5)}
data
# Trabajando con diccionarios
```

```python
s = pd.Series(data)
s
```

```python
s1 = s + 2
s1
# Aumenta 2 a cada valor de la serie
```

```python
s2 = s1 + s
s2
# Suma entre series
```
