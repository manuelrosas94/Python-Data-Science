# Traduciendo variables

```python
datos.columns
# Obtener el nombre de las columnas
```

```python
diccionario_columnas = {'LIMIT_BAL': 'Limite', 
                        'CHECKING_ACCOUNT': 'Cuenta corriente', 
                        'EDUCATION': 'Escolaridad', 
                        'MARRIAGE': 'Estado civil', 
                        'AGE': 'Edad',
                        'BILL_AMT': 'Valor factura', 
                        'PAY_AMT': 'Valor pago', 
                        'DEFAULT': 'Moroso'
                        }
# Crear un diccionario con el nuevo nombre para las columnas
```

```python
tarjetas = datos.rename(columns = diccionario_columnas)
# Renombrar las columnas
```

```python
tarjetas.head()
```

```python
tarjetas['Cuenta corriente'].unique()
```

```python
dic_cuenta = {
    'Yes': 'Si', 
    'No': 'No'
}
# Traducir cuenta
```

```python
tarjetas['Cuenta corriente'].map(dic_cuenta)
# map hace un mapeo por la columna y reemplaza las variables con el nuevo diccionario
```

```python
tarjetas.head()
# No se actualizo el nuevo diccionario en el dataframe
```

```python
tarjetas['Cuenta corriente'] = tarjetas['Cuenta corriente'].map(dic_cuenta)
# Actualizar la columna con el nuevo valor
```

```python
tarjetas.head(10)
# Dataframe actualizado
```

```python
# Hacer el mismo procedimiento para la columna Escolaridad
```

```python
tarjetas['Escolaridad'] = tarjetas['Escolaridad'].map(dic_escolaridad)
```

# Test de hipotesis

```python
sns.lmplot(x = 'IU', y = 'Valor factura', data = tarjetas, col = 'Moroso');
```

```python
from scipy.stats import ranksums
```

H<sup>null<sup>

La distribucion de los grupos morosos y no morosos es la misma

H<sup>null<sup>

La distribucion de los grupos morosos y no morosos no es la misma

```python
moroso = tarjetas.query('Moroso == 1')['Valor factura']
```

```python
moroso
```

```python
no_moroso = tarjetas.query('Moroso == 0')['Valor factura']
```

```python
no_moroso
```

```python
resultado = ranksums(moroso, no_moroso)
resultado
# pvalue si es menor de 5% la hipotesis alternativa es aceptada, la dostribucion de los grupos moroso y no moroso no es la misma
```

```python
print("El p_value de nuestro test es {}".format(resultado.pvalue))
```

```python
# La distribucin del valor de factura entre los grupos moroso y no moroso no es igual, por lo que podemos aceptar la hiptesis alternativa.
```
